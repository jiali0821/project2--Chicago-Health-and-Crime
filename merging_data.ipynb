```python
import pandas as pd 
```

```python
#importing datasets
crime_data = "Crimes_2015_cleaned.csv"
crime_data_pd = pd.read_csv(crime_data)

ph_data="Health data.csv"
ph_data_pd = pd.read_csv(ph_data)
```

```python
crime_data_pd.head()
```

```python
ph_data_pd.head()
```

```python
#merging public health indicators with crime data based on community area
combo_data=pd.merge(crime_data_pd, ph_data_pd, on="Community Area")
combo_data.head()
```

```python
combo_data
```

```python
##exporting combo data
combo_data.to_csv("combo_data.csv", index=False, header =True)
```

```python
total_crime = combo_data.groupby(["Community Area"]).count()
total_crime
```
